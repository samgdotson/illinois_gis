```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob 
import os
```

```python
path = 'nrel_psm_data/*.csv'
files = glob.glob(path)
files.sort()
files
```

```python
# frames = []
for file in files:
    fname = file.split('/')[1].strip('.csv')
    print(fname)
    df = pd.read_csv(file,  
                     usecols=['Year','Month', 'Day', 'Hour', 'Minute', 'GHI'],)
    cols=['Year','Month', 'Day', 'Hour', 'Minute']
    df['time'] = pd.to_datetime(df[cols])
    df.drop(columns=cols, inplace=True)
    df.set_index('time', inplace=True)
    df.to_csv(f'psm_transformed/{fname}_transformed.csv')
#     frames.append(df)
# full_df = pd.concat(frames)
# full_df
```
